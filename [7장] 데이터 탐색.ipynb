```python
import pandas as pd
#파일 경로를 찾고 변수 file_path에 저장 exan=m_sample.csv
file_path = 'C:/Users/duswn/Desktop/연구실/책교정/exam_sample.csv'

#read_csv()함수로 데이터프레임 변환
df = pd.read_csv(file_path)
df

#math변수에 대해서 3개 동일한 길이로 범주형 변수로 만들어보겠습니다
#범주 구간은 Categories(3, interval[float64]:[(19.93,43.333]<(43.333, 66.667]<(66.667,90.0]]
df1 = pd.cut(df.math,3)
df1
```

```python
#math변수에 3개 구간의 범주를 그룹 함수를 추가해서 각 범주의 그룹별로 agg()함수인 개수(count), 평균(mean), 표준편차(std), 최소값(min), 최대값(max)값을 계산해봅니다
df2 = df.math.groupby(df1)
df3 = df2.agg(['count','mean', 'std','min','max'])
df3
```

```python
#pd.qcut()함수를 사용 math에 대해서 각 범주 구간 별로 동일하게 3개의 개수를 갖도록 범주를 만들 수도 있습니다 이때 labels=False로 설정하여 label 이 0,1,2구간을 0부터 순차적으로 1씩 증가하게 합니다
df4 = pd.qcut(df.math, 3, labels=False)
```

```python
#아래처럼 labes=np.arange(3,0,-1)로 직접 지정하면 label이 3,2,1로 3부터 1씩 줄어드는 순서로 할당됩니다. 위의 label과 정반대로 할당이 되었습니다
import numpy as np
df5 = pd.qcut(df.math, 3, labels=np.arange(3,0,-1))
df5
```

```python
#[3<2<1]순서로 동일 개수로 나눈 3개의 그룹별 통계량을 계산할 수 있습니다
df6 = df.math.groupby(df5)
df7 = df6.agg(['count','mean','std','min','max'])
df7
```

```python
#판다스 라이브러리 불러오기
import pandas as pd
#파일 경로를 찾고 변수 file_ppath 에 저장
file_path = 'C:/Users/duswn/Desktop/연구실/책교정/timeseries.csv'
#read_csv()함수로 데이터프레임 변환
df = pd.read_csv(file_path)
#현재 날짜를 나타내는 Data 칼럼은 문자형(object)이므로 to_datetime()함수를 이용해서 Date 칼럼을 시계열 객체(Timestamp)로 변환
df.info()
```

```python
df['new_Date'] = pd.to_datetime(df['Date'])
print(df.head())
print('\n')
print(df.info())
print('\n')
print(type(df['new_Date'][0]))
```

```python
#기존 Date 열을 지우고 new_Date를 인덱스로 지정
#데이터 요약 정보를 보면, 인덱스가 DatetiemIndex 변경되었고 2015년 07월 02일에서 2019년 06월 26일 사이에 5개 날짜가 있습니다
df.drop('Date', axis=1, inplace=True)
df.set_index('new_Date', inplace=True)
print(df.head())
print('\n')
print(df.info())
```

```python
#Timestamp와 Period의 차이 확인, Timestamp를 Period로 변환
dates = ['2020-01-01', '2020-03-01', '2021-09-01']
ts_dates = pd.to_datetime(dates)
print(ts_dates)
print('\n')
```

```python
#Timestamp를 Period 변환
pr_day = ts_dates.to_period(freq='D')
print(pr_day)
pr_month = ts_dates.to_period(freq='M')
print(pr_month)
pr_year = ts_dates.to_period(freq='A')
print(pr_year)
```

```python
#시계열 활용
#날짜 데이터 분리: dt.year, dt.month, dt.day
#판다스 라이브러리 불러오기
import pandas as pd
#파일 경로를 찾고 변수 file_path에 저장
file_path = 'C:/Users/duswn/Desktop/연구실/책교정/timeseries.csv'
#read_csv()함수로 데이터프레임 변환
df = pd.read_csv(file_path)
df.info()
df['new_Date'] = pd.to_datetime(df['Date'])
print(df.head())
print('\n')
```

```python
df['Year'] = df['new_Date'].dt.year
df['Month'] = df['new_Date'].dt.year
df['Day'] = df['new_Date'].dt.day
df.head()
```

```python
#to_period()함수를 이용해 표기 변경
df['Date_yr'] = df['new_Date'].dt.to_period(freq='A')
df['Date_m'] = df['new_Date'].dt.to_period(freq='M')
df.head()
```

```python
#날짜 인덱스 지정
df.set_index('new_Date', inplace = True)
df.head()
#날짜 인덱싱
print(df.loc['2015-07'])
print('\n')
print(df['2019-06'])
print('\n')
print(df['2018-06-25':'2018-06-20'])
```

```python
#오늘 날짜와 차리 열 추가
today = pd.to_datetime('2020-03-18')
df['time_diff'] = today-df.index
df.head()
```
