# 4. 중복 데이터
187p

```python
import pandas as pd
#파일 경로를 찾고 변수 file_path에 저장 dup_sample.csv
file_path = 'C:/Users/duswn/Desktop/연구실/책교정/diabetes.csv'

#read_csv()함수로 데이터프레임 변환
#인덱스 1데이터와 5,7이 중복되었다
df = pd.read_csv(file_path)
df
```

```python
#duplicated()메소드로 중복 데이터 찾기
# DataFrame.duplicated()는 True, False의 boolean형태의 Series 반환
df.duplicated(['Pregnancies'])
```

```python
df.duplicated(['Pregnancies','BloodPressure'])
```

188p 중복 시작과 끝 확인

```python
#중복이 있으면 처음이나 끝에 무엇을 남길지 확인
# keep = 'first'가 default이며, 중복값이 있으면 첫 번째 값을 duplicated 여부를 False로 반나머지 중복값에 대해서는 True를 반환
df.duplicated(['Pregnancies'], keep='first')
```

```python
# keep='last'는 중복값이 있으면 첫 번째 값을 duplicated 여부를 True로 반환 나머지 중복값에 대해서는 False를 반환
df.duplicated(['Pregnancies'], keep='last')
```

```python
#keep = False는 처음이나 끝값인지 여부는 고려를 안하고 중복이면 무조건 True를 반환
df.duplicated(['Pregnancies'], keep = False)
```

189p 중복제거 

```python
#drop_duplicateds()는 중복값을 keep='first','last', False argument에 따라 유일한 1개의 key값만 남기고 나머지는 중복 제거
df.drop_duplicates(['Pregnancies'], keep='first')
```

```python
df.drop_duplicates(['Pregnancies'], keep='last')
```

```python
df.drop_duplicates(['Pregnancies'], keep=False)
```
